## Multi Agent Chat

```python
openai = OpenAI(api_key=openai_api_key)
antropic = OpenAI(base_url=antropic_url, api_key=anthropic_api_key)
```

```python
gpt_system_prompt = "You are a confident and aggressive debater. Challenge every argument, find flaws, and present strong counterpoints. Be sharp and slightly provocative, but stay logical and do not give false information."
claude_system_prompt = "You are a deeply curious thinker. Ask smart, challenging questions that expose assumptions and push the discussion deeper. Focus on understanding, not arguing."
```

```python
gpt_message = ["Hi"]
claude_message = ["Hello :), is AI replacing human in Tech field?"]
```

```python
def call_gpt():
    messages = [
        {
            "role": "system",
            "content": gpt_system_prompt
        }
    ]

    for gpt, claude in zip(gpt_message, claude_message):
        messages.append({"role":"assistant", "content": gpt})
        messages.append({"role":"user", "content": claude})
    
    response = openai.chat.completions.create(
        model="gpt-5.4-nano",
        messages=messages
    )

    return response.choices[0].message.content
```

```python
def call_claude():
    messages = [{"role": "system", "content": claude_system_prompt}]
    for gpt,claude in zip(gpt_message, claude_message):
        messages.append({"role":"user", "content": gpt})
        messages.append({"role":"assistant", "content": claude})
    response = antropic.chat.completions.create(
        model="claude-haiku-4-5-20251001",
        messages=messages
    )
    return response.choices[0].message.content
```

```python
call_gpt()
```

```python
call_claude()
```

```python
gpt_messages = ["Hi there!"]
claude_messages = ["Hello there! is AI replacing human in Tech field?"]

display(Markdown(f"###GPT:\n{gpt_messages[0]}\n"))
display(Markdown(f"###Claude:\n{claude_messages[0]}\n"))

for i in range(5):
    gpt_response = call_gpt()
    claude_response = call_claude()
    gpt_messages.append(gpt_response)
    claude_messages.append(claude_response)
    display(Markdown(f"###GPT:\n{gpt_response}\n"))
    display(Markdown(f"###Claude:\n{claude_response}\n"))
```

```python
import os 
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import display, Markdown
load_dotenv(override=True)
```

```python
import gradio as gr
```

```python
groq_api_key = os.getenv("GROQ_API_KEY")
groq_base_url = "https://api.groq.com/openai/v1"
```

```python
groq = OpenAI(base_url=groq_base_url, api_key=groq_api_key)
groq_model = "openai/gpt-oss-120b"
```

```python
system_prompt = "You are a helpful assistant replies in funny ways."

def message_groq(prompt):
    messages = [{"role": "system", "content": system_prompt}, {"role": "user", "content": prompt}]
    response = groq.chat.completions.create(
        model=groq_model,
        messages=messages,
    )
    return response.choices[0].message.content
    
```

```python
response = message_groq("What is today's date?")
display(Markdown(f"**Groq Response:** {response}"))
```
