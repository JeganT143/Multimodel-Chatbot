```python
import os 
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import display, Markdown
load_dotenv(override=True)
```

```python
openai_api_key = os.getenv("OPENAI_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")
antropic_api_key = os.getenv("ANTROPIC_API_KEY")
grok_api_key = os.getenv("GROK_API_KEY")
groq_api_key = os.getenv("GROQ_API_KEY")
open_router_api_key = os.getenv("OPEN_ROUTE_API_KEY")
ollama_host = os.getenv("OLLAMA_HOST")
ollama_default_model = os.getenv("OLLAMA_DEFAULT_MODEL")
```

```python
if openai_api_key:
    print("OpenAI API Key is set.")
if google_api_key:
    print("Google API Key is set.")
if antropic_api_key:
    print("Anthropic API Key is set.")
if grok_api_key:
    print("Grok API Key is set.")
if groq_api_key:
    print("Groq API Key is set.")
if open_router_api_key:
    print("Open Router API Key is set.")
if ollama_host:
    print("Ollama Host is set.")
if ollama_default_model:
    print("Ollama Default Model is set.")
```

```python
openai = OpenAI(api_key=openai_api_key)

antropic_url = "https://api.anthropic.com/v1/"
gemini_url = "https://generativelanguage.googleapis.com/vibeta/openai/"
groq_url = "https://api.groq.com/openai/v1/"
grok_url = "https://api.x.ai/v1/"
open_router_url = "https://openrouter.ai/api/v1"
ollama_url = "http://localhost:11434/v1"

antropic = OpenAI(base_url=antropic_url, api_key=antropic_api_key)
gemini = OpenAI(base_url=gemini_url, api_key=google_api_key)
groq = OpenAI(base_url=groq_url, api_key=groq_api_key)
grok = OpenAI(base_url=grok_url, api_key=grok_api_key)
open_router = OpenAI(base_url=open_router_url, api_key=open_router_api_key)
ollama = OpenAI(base_url=ollama_url, api_key="ollama")  
```

```python
tell_a_joke =[
    {
        "role": "system",
        "content": "You are a helpful assistant that tells jokes."
    },
    {
        "role": "user",
        "content": "Tell a joke in one sentence to a AI engineer"
    }
]
```

```python
response = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=tell_a_joke
)
display(Markdown(f"**OpenAI Response:** {response.choices[0].message.content}"))
```

```python
response = antropic.chat.completions.create(
    model="claude-3-haiku-20240307",
    messages=tell_a_joke
)
display(Markdown(f"**Anthropic Response:** {response.choices[0].message.content}"))
```

### Training vs Inference time scaling

```python
easy_buzzle = [
    {
        "role":"user",
        "content":"What is the probability of getting 3 heads in a row when flipping a fair coin one by time? and roll a 6 sided die and getting a 4? write only the final answer without any explanation"
    }
]
```

```python
response_openai = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=easy_buzzle,
    reasoning_effort="high"
)

display(Markdown(f"**OpenAI Response to Easy Buzzle:** {response_openai.choices[0].message.content}"))
```

```python
hard = """Find all integer solutions to:
x2−y2=2024
"""

hard_buzzle = [
    {
        "role":"user",
        "content": hard
    }   
]

response_openai_hard = openai.chat.completions.create(
    model="gpt-4-turbo",
    messages=hard_buzzle,
    reasoning_effort="none"
)
display(Markdown(f"**OpenAI Response to Hard Buzzle:** {response_openai_hard.choices[0].message.content}"))
```

```python
response_anthropic_hard = antropic.chat.completions.create(
    model="claude-3-opus-20240229",
    messages=hard_buzzle,
)
display(Markdown(f"**Anthropic Response to Hard Buzzle:** {response_anthropic_hard.choices[0].message.content}"))
```

```python
groq_response = groq.chat.completions.create(
    model="llama3-70b-8192",
    messages=hard_buzzle,
)
display(Markdown(f"**Groq Response to Hard Buzzle:** {groq_response.choices[0].message.content}"))
```

```python
ollama_response = ollama.chat.completions.create(
    model="gpt-oss:120b-cloud",
    messages=hard_buzzle,
)
display(Markdown(f"**Ollama Response to Hard Buzzle:** {ollama_response.choices[0].message.content}"))
```
